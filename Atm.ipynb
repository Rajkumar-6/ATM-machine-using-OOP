```python
import time
class Atm:
  def __init__(self):
    self.pin=1234
    self.balance=500
    self.account_no="12345"
    self.phn_no="9390"
    self.menu()
  def menu(self):
     user_input=input("""
     Welcome to ATM
  Hello!How would you like to proceed?
  1.Enter 1 to create pin
  2.Enter 2 to Deposit amount
  3.Enter 3 to Balance Enquiry
  4.Enter 4 to Withdraw amount
  5.Enter 5 to change pin
  6.Enter 6 to Exit
""")
     if user_input=="1":
       self.create_pin()
     elif user_input=="2":
        self.deposit()
     elif user_input=="3":
        self.Balance_Enquiry()
     elif user_input=="4":
        self.Withdraw()
     elif user_input=="5":
        self.change_pin()
     else:
        self.Exit()

  def create_pin(self):
    act_number=input("enter account number")
    phone=input("enter linked phone number:")
    if act_number==self.account_no and phone==self.phn_no:
      new_pin=int(input("enter new pin:"))
      print("pin created successfully")
      self.pin=new_pin
    else:
      print("Invalid Details")
   
    self.menu()
  def deposit(self):
    input_pin=int(input("enter your Pin"))
    if input_pin==self.pin:
      amount=int(input("enter your Deposit amount:"))
      self.balance=self.balance+amount
      print("amount deposited successfully")
      print("would you want show total balance:")
      option=input("""
         1.Yes
         2.No
      
      """)
      if option=="1":
        print("Total balance:",self.balance)
      else:
        print("Thank you!")
    else:
      print("wrong details")
   
    self.menu()
  def Balance_Enquiry(self):
    input_pin=int(input("enter your Pin"))
    if input_pin==self.pin:
      print("your current balance is",self.balance)
    else:
      print("Invalid Pin!!")
    self.menu()
  def Withdraw(self):
    input_pin=int(input("enter your Pin"))
    if input_pin==self.pin:
      amount=int(input("enter the amount you want:"))
      if amount>self.balance:
        print("Insufficient balance")
      else:
        print("please collect your money")
        time.sleep(5)
        self.balance=self.balance-amount
        print("would you want show total balance:")
        option=input("""
         1.Yes
         2.No
      
      """)
      if option=="1":
        print("Total balance:",self.balance)
      else:
        print("Thank you!")
    self.menu()

  def change_pin(self):
    act_number=input("enter account number:")
    phone=input("enter linked phone number:")
    if act_number==self.account_no and phone==self.phn_no:
      new_pin=int(input("enter New Pin"))
      print("pin changed Successfully!!")
    else:
      print("wrong details")
    self.menu()

  def Exit(self):
    print("Thank You!! Visit Again ")



sbi=Atm()


```
